### Solution 2

The horizontal lines correspond to metabolites (mostly co-factors) that appear in many reactions.

```python
import networkx as nx
def reactions_to_network(reactions):
    edges = list()
    for reaction in reactions:
        substrates = list(reaction.reactants)
        for substrate in substrates:
            products = list(reaction.products)
            for product in products:
                if reaction.reversibility:
                    edges.append((product, substrate, dict(reaction=reaction)))
                    edges.append((substrate, product, dict(reaction=reaction)))
                elif reaction.lower_bound > 0:
                    edges.append((substrate, product, dict(reaction=reaction)))
                else:
                    edges.append((product, substrate, dict(reaction=reaction)))

    multi_graph = nx.MultiDiGraph(edges)
    return multi_graph
```

```python
network = reactions_to_network(model.reactions)
```

These metabolites are the reason metabolic network look like giant hairballs.

```python
nx.draw_spring(network)
```
